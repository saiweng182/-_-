```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings('ignore')

pd.set_option('max_columns',1000) 
pd.set_option('max_row',300)  
#pd.set_option(‘display.float_format’, lambda x: ‘%.5f’ % x)
```

```python
#读取
train_ = pd.read_csv(r'D:\\BaiduNetdiskDownload\\Data\\二手车\\train_2.csv')
test_ = pd.read_csv(r'D:\\BaiduNetdiskDownload\\Data\\二手车\\test_2.csv')
```

```python
#特征工程
#分类特征['brand','bodyType','fuelType','gearbox','notRepairedDamage',]
#回归特征['power','kilometer']
#未使用特征['regionCode','name','model']

#构造特征
def feature_work(data):
    global gearbox_dict,bodyType_dict,brand_dict,model_dict
    #使用时间
    data['use_day'] = ((data.creat_year - data.reg_year)*12 + (data.creat_month - data.reg_month))*30 + (data.creat_day - data.reg_day)
    data['use_month'] = (data.creat_year - data.reg_year)*12 + (data.creat_month - data.reg_month)
    data['use_year'] = data.creat_year - data.reg_year
    
    data['power'] = np.log(data['power']//10 + 1)
    data['power'] = (data['power'] - data['power'].min())/(data['power'].max() - data['power'].min())
    data['kilometer'] = (data['kilometer'] - data['kilometer'].min())/(data['kilometer'].max() - data['kilometer'].min())
    
    if 'price' in data.columns:
        data['price'] = np.log1p(data['price'])
    
    
    Poly_F = ['power', 'kilometer','v_3','v_8','v_10','v_12','v_14','use_day','use_month','use_year','reg_year']
    ploy = PolynomialFeatures(degree = 2)
    data_P = pd.DataFrame(ploy.fit_transform(data[Poly_F]),columns=ploy.get_feature_names())
    index_P = ploy.get_feature_names()[1:]
    data = pd.concat([data,data_P[index_P]],axis=1)

    return (data,index_P)

train_,index_P = feature_work(train_)
test_,_ = feature_work(test_)
```
